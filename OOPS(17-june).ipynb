```python
class student():
    def add_data(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print("Name" , self.name)
        print("Age", self.age)

    def voting(self):
        if self.age >= 18:
            print(self.name, "CAN VOTE")
        else:
            print(self.name, "CAN  NOT VOTE")
```

```python
rahul = student()
rahul.add_data("vivek",19)
rahul.voting()
```

```python
class bankaccount:
    def __init__ (self, name, balance):
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        self.balance = self.balance + amount

    def withdraw(self, amount):
        self.balance = self.balance - amount
    
    def show(self):
        print("Name" , self.name)
        print("Balance", self.balance)
```

```python
rahul = bankaccount("Rahul", 100000)
```

```python
rahul.show()
```

```python
rahul.deposit(2000)
```

```python
rahul.withdraw(1000)
```

```python
rahul.show()
```
