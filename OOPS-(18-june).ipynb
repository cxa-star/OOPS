```python
class parent:
    def display(self):
        print("I am parent class")
```

```python
# super class ... parent ke display ko call karega

class child(parent):
    def display(self):
        super().display()
        print("I am child class")
```

```python
ch = child()
```

```python
ch.display()
```

# Single level Inheritance

```python
class teacher:
    def teaches(self):
        print("Teacher is teaching OOPS!")

class students(teacher):
    def sound(self):
        print("Students are making noise")

s1 = students()

s1.teaches()
s1.sound()
```

# Multi Level Inheritance

```python
class teacher:
    def teaches(self):
        print("Teacher is teaching OOPS!")

class students(teacher):
    def sound(self):
        print("Students are making noise")

class others(students):
    def doing(self):
        print("They are doing nothing!")

s1 = others()

s1.teaches()
s1.sound()
s1.doing()
```

# Method Overriding

```python
class birds:
    def fly(self):
        print("Birds are flying!")

class penguin(birds):
    def fly(self):
        print("penguin cannot fly")

p1 = penguin()

p1.fly()
```

# super keyword

```python
class girl:
    def beautiful(self):
        print("she is beautiful")

class babygirl(girl):
    def beautiful(self):
        super().beautiful()
        print("baby girls are cute")

b1 = babygirl()
b1.beautiful()
```

# Polymorphism

```python
class dog:
    def sounds(self):
        print("dog is barking!!")

class cat:
    def sounds(self):
        print("cat meoww....")

d1 = dog()
d1.sounds()

c1 = cat()
c1.sounds()
```

```python
class duck:
    def talk(self):
        print("Quackkk,... Quackkk..!!")

class humans:
    def talk(self):
        print("dont want to talk!")

def speak(obj):
    obj.talk()

d1 = duck()
h1 = humans()

speak(d1)
speak(h1)
```

# getter method

```python
class bank:
    def __init__ (self):
        self.balance = 5000
    def get_balance(self):
        return self.__balance         # getter method
```

```python
b1 = bank()
```

```python
b1.get_balance
```

# setter method 

```python
# setter method...

class bank:
    def __init__ (self):
        self.balance = 5000

    def deposit(self, amount):
        self.balance += amount  # hides the amount
        
    def get_balance(self):
        return self.__balance

b1 = bank()

b1.deposit(200)
```

```python
b1.get_balance
```
